```python
# Using the customer dataset, perform the following:
# a. 2D Clustering based on Age and Spending Score
# b. Decide the k value 
# c. Apply K-means for k = 4 and k = 5
# d. 2D Clustering based on Annual Income and Spending Score
# e. Plot the graphs for all the 3 cases.

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## 2D Clustering based on Age and Spending Score

```python
# a. 2D Clustering based on Age and Spending Score

df = pd.read_csv('./datasets/Mall_Customers.csv', index_col='CustomerID')
df.head()
```

```python
df.columns
```

```python
x = df.loc[:,['Age','Spending Score (1-100)']].values
```

```python
# b. Decide the k value 
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, n_init=10 ,init='k-means++', random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

Let's take K as 4

```python
# c. Apply K-means for k = 4 and k = 5
model = KMeans(n_clusters = 4,n_init=10, init='k-means++', random_state=42)
ypred = model.fit_predict(x)
```

```python
plt.figure(figsize =(10,10))
plt.scatter(x[ypred == 0, 0], x[ypred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[ypred == 1, 0], x[ypred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[ypred == 2, 0], x[ypred == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[ypred == 3, 0], x[ypred == 3, 1], s = 100, c = 'magenta', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.show()
```

Now take K as 5

```python
model = KMeans(n_clusters = 5 ,n_init=10, init='k-means++', random_state=42)
ypred = model.fit_predict(x)
```

```python
plt.figure(figsize =(10,10))
plt.scatter(x[ypred == 0, 0], x[ypred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[ypred == 1, 0], x[ypred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[ypred == 2, 0], x[ypred == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[ypred == 3, 0], x[ypred == 3, 1], s = 100, c = 'magenta', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.show()
```
